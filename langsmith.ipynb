```python
from langgraph.graph import StateGraph, START, END
# from langgraph.checkpoint import Checkpoint
from typing import TypedDict, Annotated, List
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.messages import SystemMessage, HumanMessage
from dotenv import load_dotenv
from langgraph.checkpoint.memory import InMemorySaver 
from langchain_core.output_parsers import StrOutputParser
from langchain_core.prompts import PromptTemplate
```

```python
load_dotenv()
llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash")
```

```python
#define the state
class Bot(TypedDict):
    question : str
    answer : str
    token_used : int
```

```python
def question(state: Bot):
    question_text = state["question"]
    rewritten = llm.invoke(
        "Rewrite this question in a clearer way: " + question_text
    ).content
    return {"question": rewritten}
```

```python
def answer(state : Bot):
    question_text = state["question"]
    answer_text = llm.invoke(
        "Answer this question in detail: " + question_text
    ).content
    return {"answer": answer_text}
```

```python
def token(state : Bot):
    question_text = state["question"]
    answer_text = state["answer"]
    token_count = len(question_text) + len(answer_text)
    return {"token_count": token_count}
```

```python
graph = StateGraph(Bot)
#add the nodes
graph.add_node("question", question)
graph.add_node("answer", answer)
graph.add_node("token", token)

#add the edges
graph.add_edge(START, "question")
graph.add_edge("question", "answer")
graph.add_edge("answer", "token")
graph.add_edge("token", END)

checkpoint = InMemorySaver()
workflow = graph.compile(checkpointer=checkpoint)
```

```python
config1 = {"configurable": {"thread_id": "1"}}
workflow.invoke({'question' : 'ai vs india education'}, config=config1)
```
